```python
# changer les valeurs de 'REMARQUABLE' pour que ca soit plus explicite

condition = df.REMARQUABLE == 1

df[condition].shape
```

```python
# changer la valeur 1 pour le champ remarquable en la valeur 'oui'
condition = df.REMARQUABLE == 1
df.loc[condition, 'REMARQUABLE'] = 'oui'
```

```python
df['REMARQUABLE'].value_counts(dropna = False)
```

```python
# changer la valeur 0 pour remarquable en 'non'

condition = df.REMARQUABLE == 0
df.loc[condition, 'REMARQUABLE'] = 'non'

print(df.REMARQUABLE.value_counts(dropna= False))


```

```python
# NaN > non

condition = df.REMARQUABLE.isnull()
df.loc[condition, 'REMARQUABLE'] = 'non'
print(df.REMARQUABLE.value_counts(dropna= False))
```
