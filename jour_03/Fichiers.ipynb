```python
# 
import os
from pathlib import Path 
import glob

def extraire_information(contents, fichier,repertoire):
    dico = {
        "contenu" : contents,
        "longueur": len(contents),
        "fichier" : fichier.split('/')[-1],
        "repertoire": repertoire.split('/')[-1]
    }
    return dico


DATA_PATH = '/home/alexis/py918/data/bbc/*' 
liste_fichiers_01 = glob.glob(DATA_PATH )    

liste_directories = []
for elt in liste_fichiers_01:
    if os.path.isdir(elt):
        liste_directories.append(elt)

# equivalent a
# liste_directories = [ elt for elt in liste_fichiers_01  if os.path.isdir(elt)  ]
        
liste_textes = []

for repertoire in liste_directories:
    liste_fichiers = glob.glob(repertoire + "/*")
    print("{} a {} fichiers  ".format( repertoire, len(liste_fichiers)  ))

    for fichier in liste_fichiers:
        try: 
            contents = Path(fichier).read_text() 
            dico     = extraire_information(contents, fichier, repertoire)
            liste_textes.append(dico)
        except:
            print("fichier: {} ".format(fichier))
```

# stats

On a la liste des dictionnaires. Chaque dico correspond a un fichier.

On veut maintenant connaitre pour chaque repertoire : la longueur  moyenne des textes
et mettre cette info dans un nouveau dictionnaire.







```python
# list comprehension
repertoires_01 = list(set([ dico['repertoire']  for dico in liste_textes  ]))
```

```python
# equivalent a
repertoires_02 = []

for dico in liste_textes:
    repertoires_02.append(dico['repertoire'])

repertoires_02 = list(set(repertoires_02))
repertoires_02
```

```python
assert repertoires_01  == repertoires_02
```

```python
stats = {}

# initialise

for repertoire in repertoires_01:
    stats[repertoire] = []
    
stats  
```

```python
import numpy as np
for dico in liste_textes:
    repertoire = dico['repertoire']
    longueur = dico['longueur']
    stats[repertoire].append(longueur)    

    
moyenne = {}    
for repertoire in repertoires_01:
    moyenne[repertoire] = np.mean(stats[repertoire])
    
```

# Ecrire les infos dans un fichier csv

* ouvrir un nouveau fichier pour l'ecriture
* boucler sur chaque dico
* ecrire les elements du dico dans le fichier en séparant par une virgule et en mettant le contenu entre ""



```python
#         "contenu" : contents,
#         "longueur": len(contents),
#         "fichier" : fichier.split('/')[-1],
#         "repertoire": repertoire.split('/')[-1]

with open('bbc.csv', 'a') as f:

    en_tete = '"longueur","repertoire","fichier","contenu"\n'
    f.write(en_tete)
    
    for dico in liste_textes:
        # dans le contenu remplacer les simples quotes ' par des " quotes
        contenu = dico['contenu'].replace('"',"'")
        # dans le contenu remplacer les retour lignes par un espace
        contenu = contenu.replace('\n'," ")
        
        # contruire la ligne au format csv
        
        ligne = '"{}","{}","{}","{}"\n'.format(
                        dico['longueur'], 
                        dico['repertoire'], 
                        dico['fichier'],
                        contenu
        )
        # ecrire la ligne dans le fichier
        f.write(ligne)
```
