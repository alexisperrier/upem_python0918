# quelques visualisations





```python
import pandas as pd

arbres = pd.read_csv('arbres_et_superficies.csv')
arbres.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1,1, figsize = (6,6))
plt.scatter(arbres.nombre_arbres, arbres.SURFACE, label = 'arbres', color = "red")
plt.xlabel("Nombre d'arbres")
plt.ylabel("Superficie")
plt.title("Superficie vs Arbres")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

np.arange(5)
```

```python
arbres.sort_values(by = 'nombre_arbres', ascending = False, inplace = True)

y_pos = np.arange(len(arbres.C_ARINSEE))
plt.bar(y_pos, arbres.nombre_arbres, align='center', alpha=0.5)
plt.xticks(y_pos, arbres.C_ARINSEE)

plt.xlabel('Arrdt')
plt.ylabel("Nombre d'arbres")
plt.xticks(rotation=70)
plt.show()



```

```python
arbres.sort_values(by = 'SURFACE', ascending = False, inplace = True)

y_pos = np.arange(len(arbres.C_ARINSEE))
plt.bar(y_pos, arbres.SURFACE, align='center', alpha=0.5)
plt.xticks(y_pos, arbres.C_ARINSEE)

plt.xlabel('Arrdt')
plt.ylabel("Superficie")
plt.xticks(rotation=70)
plt.show()


```

```python
arbres['ratio'] = arbres.nombre_arbres / arbres.SURFACE
arbres.head()
```

```python
arbres.sort_values(by = 'ratio', ascending = False, inplace = True)

y_pos = np.arange(len(arbres.C_ARINSEE))
plt.bar(y_pos, arbres.ratio, align='center', alpha=0.5)
plt.xticks(y_pos, arbres.C_ARINSEE)

plt.xlabel('Arrdt')
plt.ylabel("ratio")
plt.xticks(rotation=70)
plt.show()
```
