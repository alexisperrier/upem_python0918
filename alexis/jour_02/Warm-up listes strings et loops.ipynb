# 1. liste des mots

En utilisant split(), construire la liste des mots du texte

```python
liste_des_mots = texte_03.split(' ')
liste_des_mots[0:30]
```

```python
# enlever les ''


liste_des_mots_01 = liste_des_mots.copy()

while '' in liste_des_mots_01: 
    liste_des_mots_01.remove('')

liste_des_mots_01[:30]
```

# 2. Ponctuation

parcourir la liste des mots et 
* enlever la ponctuation 
* mettre en minuscule

avec 

* string.punctuation
* lower()



```python
import string

liste_des_mots_sans_ponctuation = []

for mot in liste_des_mots_01:
    
    for signe in string.punctuation:
        if signe in mot:
            mot = mot.replace(signe,' ')
    
    # a ce niveau la j'ai le mot sans les signes de ponctuations
    # je le mets dans la nouvelle liste liste_des_mots_sans_ponctuation
    mot_en_minuscule = mot.lower()
    mot_sans_espace =     mot_en_minuscule.strip()
    liste_des_mots_sans_ponctuation.append(mot_sans_espace)

print("premier elements de la liste des mots sans ponctuation: ")
print(liste_des_mots_sans_ponctuation[:20])
            
```

```python
# application a liste_des_mots_sans_ponctuation

liste_des_mots_uniques = list(set(liste_des_mots_sans_ponctuation))
```

```python
print("nombre de mots en tout {}".format(len(liste_des_mots_sans_ponctuation)))
print("nombre de mots unique {}".format(len(liste_des_mots_uniques)))

```

# 4. frequence

Calculer pour chaque mot unique le nombre d'apparition dans la liste de l'etape 2


```python
# j'ai la liste de mots uniques

texte = ''

liste_des_occurences = []

for mot in liste_des_mots_uniques:
    n = liste_des_mots_sans_ponctuation.count(mot)
    liste_des_occurences.append(n)
```

```python
# version dictionnaire

dico = {}

for mot in liste_des_mots_uniques:
    n = liste_des_mots_sans_ponctuation.count(mot)
    dico[mot] = n
    
    
dico
    

```

```python
import numpy as np


print("En moyenne les mots apparaissent {:.2f} fois".format(np.mean(liste_des_occurences)))
print("La valeur medianne {:.2f} ".format(np.median(liste_des_occurences)))

print("Le 95 ieme centile est {:.2f}".format(np.percentile(liste_des_occurences, 95)))

```

# 6. les mots les plus fréquents

Construire une liste plus restreinte de mots en ne gardant que les mots qui apparaissent plus que le median calculé auparavant.


```python
seuil = np.percentile(liste_des_occurences, 95)

liste_des_mots_frequents = []

for mot in sorted(liste_des_mots_uniques):
    n = liste_des_mots_sans_ponctuation.count(mot)
    if n > seuil:
        liste_des_mots_frequents.append(mot)
    

liste_des_mots_frequents
```

```python
for mot in liste_des_mots_frequents:
    n = liste_des_mots_sans_ponctuation.count(mot)
    print("{}: {}".format(mot, n))
```
